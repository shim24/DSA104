```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Reproducibility
np.random.seed(42)

# Generate random data: 100 samples, 5 features
X = np.random.randn(200, 5)

# Run PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Create a DataFrame for seaborn
df_pca = pd.DataFrame(
    X_pca,
    columns=["Principal Component 1", "Principal Component 2"]
)

# Optional: add a categorical variable for coloring
df_pca["Group"] = np.random.choice(["A", "B", "C"], size=len(df_pca))

# Seaborn scatter plot
plt.figure()
sns.scatterplot(
    data=df_pca,
    x="Principal Component 1",
    y="Principal Component 2",
)

plt.title("Latent life science space")
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

# Reproducibility
np.random.seed(42)

# Generate clustered data
X, y = make_blobs(
    n_samples=700,
    n_features=5,
    centers=3,
    cluster_std=4.5,
    random_state=42
)

# Standardize features (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Run PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Create DataFrame for visualization
df_pca = pd.DataFrame(
    X_pca,
    columns=["Principal Component 1", "Principal Component 2"]
)
df_pca["Cluster"] = y.astype(str)

# Plot using seaborn
plt.figure()
sns.scatterplot(
    data=df_pca,
    x="Principal Component 1",
    y="Principal Component 2",
    hue="Cluster",
    palette="deep"
)

plt.title("PCA Scatter Plot of Clustered Data")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
